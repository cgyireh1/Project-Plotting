```python
df = pd.read_csv("fuel_econ.csv")
corr_matrix = df.select_dtypes(include=['number']).corr()
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(corr_matrix, cmap='vlag_r', vmin=-1)

ax.set_xticklabels(corr_matrix.columns, rotation=45, ha='right')
ax.set_yticklabels(corr_matrix.columns, rotation=0)

plt.show()
```

```python
# Histogram for 'co2' Distribution: 
### The histogram is a skewed distribution with a long tail to the right, indicating that most of the values are concentrated in a narrow range but there are a few extreme values that are much higher. The 'co2' values in the dataset may have certain outliers.

# Histogram for 'comb' Distribution: 
### The histogram is also a skewed distribution with a long tail to the right.  This indicates that the 'comb' values may have outliers.

# Histogram of Engine Displacement: 
### The histogram is a normal distribution with a central peak and a slight skewness to the right indicating that the engine displacement values are not uniformly distributed and may follow a specific pattern.

# Heatmap plot: 
###The The correlation ranges from -1 to 1, with darker colors indicating higher correlation. The plot shows that there is a strong positive correlation between the variables 'displ' and 'comb'. There is also a strong positive correlation between the variables 'displ' and 'co2'. There is also a strong positive correlation between the variables 'comb' and 'co2'. The plot also indicates weak positive and negative correlations between other variables, 'year', 'cylinders', and 'displ'.

# Columns to drop
### If the target is ghg score, I would drop the columns 'co2' and 'comb'because these columns are highly correlated with the target variable 'ghgScore' and may introduce multicollinearity in the model. This would help in reducing the complexity of the model.
```
